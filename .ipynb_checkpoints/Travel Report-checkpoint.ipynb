```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Gathering

```python
df=pd.read_csv("Travel details dataset.csv")
```

# Data Cleaning

```python
df.dropna(inplace = True)
```

```python
df.info()
```

## Covert Column Type From Float to Integer

```python
df['Duration (days)'] = df['Duration (days)'].astype(int)
df['Traveler age'] = df['Traveler age'].astype(int)
```

## Convert Column Type From Object to Integer

```python
# Remove non-numeric characters from the 'Accommodation cost' column
df['Accommodation cost'] = df['Accommodation cost'].replace('[\$,]', '', regex=True)
df['Transportation cost'] = df['Transportation cost'].replace('[\$,]', '', regex=True)

# If you have values with 'USD' suffix, you can remove it as well
df['Accommodation cost'] = df['Accommodation cost'].replace(' USD', '', regex=True)
df['Transportation cost'] = df['Transportation cost'].replace(' USD', '', regex=True)

# Convert the column to integers
df['Accommodation cost'] = pd.to_numeric(df['Accommodation cost'], errors='coerce').astype(int)
df['Transportation cost'] = pd.to_numeric(df['Transportation cost'], errors='coerce').astype(int)
```

```python
df.info()
```

```python
df['Destination_Country'].unique()
```

## Making New column Name Destination_Country


```python
df['Destination_Country'] = df['Destination'].apply(lambda x: x.split(', ')[-1] if ', ' in x else x)
df.head()
```

```python
# Mapping dictionary for state to country conversion
state_to_country_mapping = {
    'Hawaii': 'USA',
    'New York': 'USA',
    'California': 'USA',
    'Paris': 'France',
    'Bali': 'Indonesia',
    'London': 'UK',
    'Tokyo': 'Japan',
    'Sydney': 'Australia',
    'Cape Town': 'South Africa',
    'Rio de Janeiro': 'Brazil',
    'SA':'South Africa',
    'AUS':'Australia',
    'Amsterdam': 'Netherlands',
    'Phuket': 'Thailand',
    'Santorini': 'Greece',
    'Phnom Penh': 'Cambodia',
    'Seoul': 'South Korea'
}

# Apply the mapping to the 'Destination_Country' column
df['Destination_Country'] = df['Destination_Country'].map(state_to_country_mapping).fillna(df['Destination_Country'])
```

```python
df['Traveler nationality'].unique()
```

```python
# Mapping dictionary for country name to nationality conversion
country_to_nationality_mapping = {
    'American': 'American',
    'Canadian': 'Canadian',
    'Korean': 'Korean',
    'British': 'British',
    'Vietnamese': 'Vietnamese',
    'Australian': 'Australian',
    'Brazilian': 'Brazilian',
    'Dutch': 'Dutch',
    'Emirati': 'Emirati',
    'Mexican': 'Mexican',
    'Spanish': 'Spanish',
    'Chinese': 'Chinese',
    'German': 'German',
    'Moroccan': 'Moroccan',
    'Scottish': 'Scottish',
    'Japanese': 'Japanese',
    'Italian': 'Italian',
    'Indian': 'Indian',
    'South Korean': 'Korean',  # Assuming South Korean is the same as Korean
    'French': 'French',
    'South African': 'South African',
    'Taiwanese': 'Taiwanese',
    'Indonesian': 'Indonesian',
    'USA': 'American',
    'Canada': 'Canadian',
    'South Korea': 'Korean',
    'UK': 'British',
    'China': 'Chinese',
    'Taiwan': 'Taiwanese',
    'Japan': 'Japanese',
    'Spain': 'Spanish',
    'Brazil': 'Brazilian',
    'Germany': 'German',
    'Hong Kong': 'Hong Kong',  # Assuming Hong Kong is the same as Chinese
    'United Kingdom': 'British',
    'Singapore': 'Singaporean',  # Assuming Singaporean is the same as Singapore
    'Italy': 'Italian',
    'Greece': 'Greek',  # Assuming Greek is the same as Greece
    'United Arab Emirates': 'Emirati',
    'Cambodia': 'Cambodian',
    'New Zealander': 'New Zealander',
   
}
df['Traveler nationality']=df['Traveler nationality'].map(country_to_nationality_mapping).fillna(df['Traveler nationality'])
```

# Exploratory Data Analysis

## Analysis of country which was visited mosted and Traveller Nationality

```python
count=df['Destination_Country'].value_counts()
colors=sns.color_palette("husl",n_colors=len(count))
plt.figure(figsize=(12,6))
count.plot(kind='bar', color=colors)
plt.title('Distribution of Destination Countries')
plt.xlabel('Destination Country')
plt.ylabel('Count')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
df['Traveler nationality'].unique()
```

```python
traveller_count=df['Traveler nationality'].value_counts()
tcolor=sns.color_palette("husl",n_colors=len(traveller_count))
plt.figure(figsize=(12,6))
traveller_count.plot(kind='bar' ,color=tcolor)
plt.title('Distribution of Travellers')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45,ha='right')
plt.show()
```
